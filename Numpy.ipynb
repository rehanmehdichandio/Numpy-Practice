###What is Numpy?
NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.


At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types

### Numpy Arrays Vs Python Sequences

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Pythonâ€™s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

###Braodcasting

The term boradcasting describes how numpy treats arrays with differnt shapes during arithmatic operations.
The smaller array is "broadcast"  accross the larger array so that they have compatile shapes

```python
#same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)
print(a)
print(b)
```

```python
#diff shape
c = np.arange(6).reshape(2,3)
d = np.arange(3).reshape(1,3)

print(c)
print(d)

print(c+d)
```

###Broadcasting rules


1.   **Make two arrays the same number of dimensions**

*   If the numbers of dimensions of the two arrays are different, add bew dimensions with size 1 to the head of the array with smaller dimension.


2.   **Make each dimension of the two arrays the same size**


*   If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size ofthe other array
*   If there is a dimension whose size is not 1 in either of the two arrays, it cannot be boradcasted, and an error raied

* Used in vectorization

```python
def categorical_cross_entropy(actual, predicted):
    epsilon = 1e-15 # Small value to prevent log(0)
    predicted = np.clip(predicted, epsilon, 1 - epsilon)
    loss = -np.sum(actual * np.log(predicted))
    return loss

# Example usage:
# Let's say we have 3 classes
actual_labels = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]]) # One-hot encoded actual labels
predicted_probabilities = np.array([[0.1, 0.8, 0.1], [0.9, 0.05, 0.05], [0.2, 0.2, 0.6]]) # Predicted probabilities

print("Actual Labels:")
print(actual_labels)
print("\nPredicted Probabilities:")
print(predicted_probabilities)

loss_value = categorical_cross_entropy(actual_labels, predicted_probabilities)
print(f"\nCategorical Cross-Entropy Loss: {loss_value}")
```

```python
import numpy as np

def binary_cross_entropy(actual, predicted):
    epsilon = 1e-15 # Small value to prevent log(0) or log(1)
    predicted = np.clip(predicted, epsilon, 1 - epsilon)
    loss = -np.mean(actual * np.log(predicted) + (1 - actual) * np.log(1 - predicted))
    return loss

# Example usage:
# Binary classification problem (e.g., predicting 0 or 1)
actual_labels_binary = np.array([1, 0, 1, 1, 0]) # Actual binary labels
predicted_probabilities_binary = np.array([0.9, 0.1, 0.8, 0.6, 0.3]) # Predicted probabilities for class 1

print("Actual Labels (Binary):")
print(actual_labels_binary)
print("\nPredicted Probabilities (Binary):")
print(predicted_probabilities_binary)

loss_value_binary = binary_cross_entropy(actual_labels_binary, predicted_probabilities_binary)
print(f"\nBinary Cross-Entropy Loss: {loss_value_binary}")
```
